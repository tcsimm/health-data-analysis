```python
# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/thomassimmons/c/Health/data/cleaned_health.csv")
```

```python
# Distribution of percentage
plt.figure(figsize=(10, 5))
sns.histplot(df["percentage"], bins=40, kde=True)
plt.title("Distribution of Percentage")
plt.xlabel("Percentage")
plt.ylabel("Count")
plt.show()
```

```python
# Percentage by year
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="year", y="percentage")
plt.title("Percentage by Year")
plt.xticks(rotation=45)
plt.show()
```

```python
top_outcomes = df["outcome_(or_indicator)"].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(y=top_outcomes.index, x=top_outcomes.values)
plt.title("Top 10 Most Common Outcomes")
plt.xlabel("Count")
plt.ylabel("Outcome")
plt.show()
```

```python
# Compare percentages by grouping category
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[df["grouping_category"] == "Sex"], x="group", y="percentage")
plt.title("Percentage by Sex")
plt.show()
```

```python
# Basic correlation heatmap
sns.heatmap(df[["percentage", "ci_lower", "ci_upper"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
